```python
from bing_image_downloader.downloader import download
from PIL import Image
import pandas as pd
import matplotlib.pyplot as plt
import os, sys
import glob
import json
```

```python
with open('actors.json', 'r') as fp:
   dict_act = json.load(fp)

SIZE = 2
```

```python
dict_act[:2]
```

```python
df = pd.DataFrame(dict_act)
# Extract all names from the 'name' column
names = df['name'].tolist()
```

```python

# Iterate through each name and download 15 photos for each person
for name in names[:2]:
    dir_ = f'dataset/'
    # Create the directory if it doesn't exist
    os.makedirs(dir_, exist_ok=True)
    str_face=f'{name}'
    # Download a photo for the current name
    download(str_face,
             limit=15,
             output_dir=dir_,
             adult_filter_off=True,
             force_replace=False,
             timeout=60,
             verbose=True)
```

```python
download("horse", limit=30,
         output_dir="horse_dataset",
         adult_filter_off=True)
```

```python
dataset_path = './horse_dataset/horse/'  # Adjust the path according to your directory structure

image_files = glob.glob(os.path.join(dataset_path, '*.jpg'))  # Adjust the extension if needed

# Define the number of rows and columns for the subplot
rows, cols = 3, 5

#  Create the subplot
fig, axes = plt.subplots(rows, cols, figsize=(15, 9))

# Loop through the image files and display them in the subplot
for i, file in enumerate(image_files[:rows*cols]):
    row = i // cols
    col = i % cols
    img = Image.open(file)
    axes[row, col].imshow(img)
    axes[row, col].axis('off')  # Hide the axes

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
